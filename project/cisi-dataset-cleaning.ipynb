```python
import pandas as pd
```

```python
def get_id(ids, line):
    if '.I' in line:
        id_ = line[ line.index('I')+1 : ].strip()
        try:
            str_id = int(id_)
            ids.append(str_id)
        
        except ValueError:
            print('Bad value', id_)    

        
            
ids = []
abstracts = []
abstract = ''

with open('data/CISI.ALL') as f:
    lines = f.readlines()
    abstract_started = False
    
    for line in lines:
        get_id(ids, line)
        
        if '.W' in line:
            abstract_started = True
            
        if '.X' in line:
            abstracts.append(abstract.replace(' .W', ''))
            abstract = ''
            abstract_started = False
            
        if abstract_started:
            abstract = abstract + ' ' + line.strip()
```

```python
cisi_all = pd.DataFrame({'docid':ids, 'abstract':abstracts})
cisi_all.to_csv('cisi-all.csv', index=False)
```

```python
df = pd.read_csv('data/CISI.REL', sep='\t', names=['qid_', 'docid', 'unknown-1', 'unknown-2'])
df
```

```python
df['unknown-2'] = df['unknown-1']
df['unknown-1'] = df['docid']
df
```

```python
df['qid'] = df['qid_'].apply(lambda row: row.strip().split(' ')[0])
df['docid'] = df['qid_'].apply(lambda row: row.strip().split(' ')[-1])
df.drop('qid_', axis=1, inplace=True)
df
```
